```python
import pandas as pd
import numpy as np 
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./data/pokemon.csv")
data.head()
```

```python
sb.factorplot(
    x='generation', 
    data=data,
    size=5,
    aspect=1.2,
    kind='count'
).set_axis_labels('Generation', '# of Pokemon')

plt.show()
```
